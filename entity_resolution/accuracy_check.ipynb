#### Sampling Clusters (40 per alphabet):

-  Sampled the first 40 clusters per alphabet in the order by which the ER algorithm encountered them.
- This gives us a sample of $26 \times 40 = 1040$ clusters to evaluate

```python
mapper = {
    'a': [], 'b': [], 'c': [], 'd': [], 'e': [], 'f': [], 'g': [], 'h': [],
    'i': [], 'j': [], 'k': [], 'l': [], 'm': [], 'n': [], 'o': [], 'p': [],
    'q': [], 'r': [], 's': [], 't': [], 'u': [], 'v': [], 'w': [], 'x': [],
    'y': [], 'z': [] }


for key in mapper.keys():
    i = 1
    for cluster in all_clusters:
        if cluster[0] == '':
            pass
        elif cluster[0][0] == key:
            mapper[key].append(cluster)
            i += 1
        if i > 40:
            break

for key in mapper.keys():
    print(f"Sample clusters starting with the letter {key.upper()}:\n")
    for name in mapper[key]:
        print(name)
    print()

```
