```python
from rapidfuzz import process, fuzz
import math
import re



# Optional: quick alias hits (safe ones only; use word boundaries to avoid false positives)
ALIASES = {
    r"\btoi\b": "The Times of India",
    r"\bht\b": "Hindustan Times",
    r"\bet\b": "The Economic Times",
    r"\bdna\b": "DNA (Daily News and Analysis)",
    r"\btnie\b": "The New Indian Express",
    r"\ffpj\b": "Free Press Journal",
}

def _alias_hit(raw_lower: str):
    for pat, canonical in ALIASES.items():
        if re.search(pat, raw_lower):
            return canonical
    return None

def _substring_hit(raw_lower: str, names):
    for n in names:
        if n.lower() in raw_lower:
            return n
    return None

def _best_fuzzy(raw: str, choices, threshold=75):
    candidates = []
    for scorer in (fuzz.partial_token_set_ratio, fuzz.partial_token_sort_ratio, fuzz.WRatio):
        res = process.extractOne(raw, choices, scorer=scorer)
        if res:
            candidates.append(res[1])
    if not candidates:
        return (None, 0)

    # Harsh rule: take min across scorers (so weak scorer drags it down)
    final_score = min(candidates)

    if final_score >= threshold:
        # Use WRatio (or longest match) to pick final string
        best = process.extractOne(raw, choices, scorer=fuzz.WRatio)
        return best[0], final_score
    else:
        return (None, 0)

def classify_press(raw, threshold=80):
    # handle nulls
    if raw is None or (isinstance(raw, float) and math.isnan(raw)) or (isinstance(raw, str) and not raw.strip()):
        return {"level": "unknown", "match": None, "score": 0, "method": "empty"}

    raw_str = str(raw)
    raw_lower = raw_str.lower()

    # 0) alias fast-path
    alias = _alias_hit(raw_lower)
    if alias:
        # decide national/local by membership
        level = "national" if alias in national_press else "local" if alias in local_press else "unknown"
        return {"level": level, "match": alias, "score": 100, "method": "alias"}

    # 1) exact substring fast-paths
    
    sub = _substring_hit(raw_lower, local_press)
    if sub:
        return {"level": "local", "match": sub, "score": 100, "method": "substring"}
    sub = _substring_hit(raw_lower, national_press)
    if sub:
        return {"level": "national", "match": sub, "score": 100, "method": "substring"}

    # 2) fuzzy against both lists
    
    l_match, l_score = _best_fuzzy(raw_str, local_press)
    n_match, n_score = _best_fuzzy(raw_str, national_press)

    # pick the better bucket
    if max(n_score, l_score) >= threshold:
        if n_score <= l_score:
            return {"level": "local", "match": l_match, "score": l_score, "method": "fuzzy"}
            
        else:
            return {"level": "national", "match": n_match, "score": n_score, "method": "fuzzy"}

    return {"level": "unknown", "match": None, "score": max(n_score, l_score), "method": "low-score"}

# --- Example usage on a pandas Series ---
# df["press_level_info"] = df["raw_text_col"].apply(classify_press)
# df["press_level"] = df["press_level_info"].apply(lambda x: x["level"])
```

```python
def merge_similar_keys(data_dict, score_threshold=70):
    merged_dict = {}
    processed_keys = set()

    for key1, value1 in data_dict.items():
        if key1 in processed_keys:
            continue

        similar_keys = [key1]
        for key2 in data_dict.keys():
            if key1 != key2 and key2 not in processed_keys:
                score = hybrid_strict_score(key1, key2)
                if score > score_threshold:
                    similar_keys.append(key2)

        # Find the longest key among similar ones
        longest_key = max(similar_keys, key=len)

        # Sum the values of similar keys
        merged_value = sum(data_dict[key] for key in similar_keys)

        # Add to the merged dictionary
        merged_dict[longest_key] = merged_value

        # Mark similar keys as processed
        processed_keys.update(similar_keys)

    return merged_dict

merged_complaint_types = merge_similar_keys(a_dict)
print(merged_complaint_types)
# Fill None values in 'ComplaintType' with the previous non-None value

df['ComplaintType'].fillna(method='ffill', inplace=True)

# Create a new column 'res_ComplaintType' and initialize it
df['res_ComplaintType'] = None

for index, row in df.iterrows():
    # Calculate similarity scores with merged complaint types
    scores = {}
    for key in merged_complaint_types.keys():
        # Add a check for empty strings before calculating the score
        if isinstance(row["ComplaintType"], str) and isinstance(key, str) and row["ComplaintType"].strip() != "":
            scores[key] = hybrid_strict_score(row["ComplaintType"], key)
        else:
            scores[key] = 0 # Assign a score of 0 if ComplaintType is not a valid string

    # Find the key with the maximum score
    if scores:  # Check if scores dictionary is not empty
      max_key = max(scores, key=scores.get)
      # Assign the key with the maximum score to 'res_ComplaintType'
      df.loc[index, 'res_ComplaintType'] = max_key
    else:
      df.loc[index, 'res_ComplaintType'] = None # Assign None if no scores were calculated

# Optional: Display the updated DataFrame with the new column
# display(df[['ComplaintType', 'res_ComplaintType']].head())
```
