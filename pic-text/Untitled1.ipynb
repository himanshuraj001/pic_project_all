```python
import re

with open("1992-against-press.txt", "r", encoding="utf-8") as f:
    text = f.read()



# Updated pattern: number or letter, with optional dot and whitespace
#pattern = r'(?=\b(?:\d+\.\s*|[a-zA-Z]\.?\s*)(?:Complaint(?:s)?|Publication| Sou)\b)'
pattern = r'(?:(?<=\n)|(?<=^))(?=\b(?:\d+\.\s*|[a-zA-Z]\.?\s*)(?:Complaint(?:s)?|Publication|Suo-motu|Suo|Communication|Memorandum|Reference|Objectionable|Circulation|News-item|Attack|Regarding|BES|Amar)\b)'

# This will now split cleanly without injecting 'Complaint of' into the array
complaints = re.split(pattern, text)

# Clean and print
for i, complaint in enumerate(complaints):
    cleaned = complaint.strip()
    if cleaned:
        print(f"[Complaint {i+1}]\n{cleaned}\n")
```

```python
complaints_by_line = []
for complaint_text in complaints:
    lines = complaint_text.split('\n')
    complaints_by_line.append(lines)

if complaints_by_line:
    #print("Lines of the first complaint:")
    for line in complaints_by_line[0]:
        print(line)
```

```python
cleaned_complaints_by_line = []
for complaint_lines in complaints_by_line:
    cleaned_lines = [line for line in complaint_lines if line.strip()]
    if cleaned_lines:
        cleaned_complaints_by_line.append(cleaned_lines)

complaints_by_line = cleaned_complaints_by_line
```
